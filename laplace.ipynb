```python
import cv2 as cv
import numpy as np

path = r"C:\PPN\PythonProject1\picture.jpg"
gray = cv.imread(path, cv.IMREAD_GRAYSCALE)
blurred = cv.GaussianBlur(gray, (5, 5), sigmaX=1)
lap = cv.Laplacian(blurred, ddepth=cv.CV_64F, ksize=3)
lap_abs = np.uint8(np.clip(np.abs(lap), 0, 255))

cv.imshow("Original Image", gray)
cv.imshow("Laplace Edge Detection", lap_abs)
cv.imwrite("laplace_edge_result.jpg", lap_abs)
```
