```python
import cv2

source_img = cv2.imread(r"/2.classwork/picture.jpg", cv2.IMREAD_GRAYSCALE)
blurred_img = cv2.GaussianBlur(source_img, (5, 5), sigmaX=1.4)
low_threshold = 100
high_threshold = 200
edge_detection_result = cv2.Canny(blurred_img, low_threshold, high_threshold)
cv2.imshow("Original Image", source_img)
cv2.imshow('Edge Detection Result', edge_detection_result)
cv2.imwrite('edge_detection_output.jpg', edge_detection_result)
```
